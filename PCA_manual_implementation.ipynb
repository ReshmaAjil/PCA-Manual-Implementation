```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create dummy dataset
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])

# Step 2: Standardize the data (zero mean)
X_meaned = X - np.mean(X, axis=0)
print(X_meaned)

# Step 3: Covariance matrix
cov_matrix = np.cov(X_meaned, rowvar=False)
print(cov_matrix)

# Step 4: Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
print(eigenvalues)
print(eigenvectors)

# Step 5: Sort eigenvectors by descending eigenvalues
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]
print(eigenvalues)
print(eigenvectors)

# Step 6: Select top k eigenvectors (e.g., k=1 for 1D reduction)
k = 1
eigenvectors_subset = eigenvectors[:, :k]

# Step 7: Transform the data
X_reduced = np.dot(X_meaned, eigenvectors_subset)

print("Reduced Data:\n", X_reduced)
```

```python
# Plot original data
plt.figure(figsize=(10, 5))

# Subplot 1: Original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original Data')
plt.title('Original 2D Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.axis('equal')
plt.grid(True)

# Subplot 2: PCA Reduced (1D projection)
plt.subplot(1, 2, 2)
plt.scatter(X_reduced, np.zeros_like(X_reduced), color='red', label='PCA 1D Projection')
plt.title('PCA Reduced Data (1D)')
plt.xlabel('Principal Component 1')
plt.yticks([])  # Remove y-axis ticks for 1D visualization
plt.grid(True)

plt.tight_layout()
plt.show()

```
